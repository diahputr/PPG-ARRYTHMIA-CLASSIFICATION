```python
import pandas as pd
df = pd.read_csv('predictfitur/datareadyAFPVC.csv')
```

```python
# df.drop(['Unnamed: 0'], axis=1, inplace=True)
# df.to_csv('predictfitur/readyya.csv', index=False)
df
```

```python
x = df[['min','max','mean','med','std']]
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=109)
```

```python
"""Naive Bayes"""
```

```python
from sklearn.naive_bayes import GaussianNB
import pickle
model = GaussianNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred )
cm
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
cv = KFold(n_splits=5, random_state=3, shuffle=True)
scores = cross_val_score(model, x, y, cv=cv)
acc = (scores[0] + scores[1] + scores[2] + scores[3] + scores[4]) / 5
acc
```

```python
from sklearn.model_selection import cross_val_predict
yk_pred = cross_val_predict(model, x, y, cv=cv)

from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y, yk_pred)
print(conf_mat)
```

```python
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, classification_report
report = classification_report(y, yk_pred)
print(report)
```
